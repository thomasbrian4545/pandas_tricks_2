```python
import pandas as pd
import numpy as np
```

```python
# persiapan DF
n_rows = 40
n_cols = 5

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=tuple('ABCDE'))
df.head()
```

```python
# line plot
df.head()
```

```python
df.plot(kind='line')
```

```python
df[['A','B']].plot(kind='line')
```

```python
# bar plot
df.plot(kind='bar')
```

```python
df[['C','D']].head().plot(kind='barh')
```

```python
# area plot
df[['A', 'B']].head().plot(kind='area')
```

```python
# box plot
df.plot(kind='box')
```

```python
# histogram
df[['A', 'B']].plot(kind='hist')
```

```python
# KDE
df.plot(kind='kde')
```

```python
# scatter plot
df.plot(x='A', y='B', kind='scatter')
```
